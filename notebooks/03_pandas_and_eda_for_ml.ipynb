# ***Pandas and Exploratory Data Analysis (EDA)***

## Objective
Learn how to use Pandas for data manipulation and perform basic EDA before applying Machine Learning models.


### **1. Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### **2. Load Dataset**

```python
data = {
    "age": [45, 54, 39, 61, 48],
    "cholesterol": [233, 250, 204, 286, 210],
    "blood_pressure": [140, 130, 120, 145, 135],
    "target": [1, 0, 1, 0, 1]
}

df = pd.DataFrame(data)
df
```

### **3. Basic Data Inspection**

```python
df.head()
df.info()
df.describe()
```

### **4. Check Missing Values**

```python
df.isnull().sum()
```

### **5. Feature & Target Separation**

```python
X = df.drop("target", axis=1)
y = df["target"]

X, y
```

### **6. Simple Visualization (EDA)**

```python
sns.pairplot(df, hue="target")
plt.show()
```

### **7. Correlation Analysis**

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

## Key Learnings
- Using Pandas DataFrame for data handling
- Inspecting data using info() and describe()
- Checking missing values
- Basic EDA using visualization
- Understanding feature relationships before ML
