# ***Python Basics for Machine Learning***

## *Objective*
Learn core Python concepts required for Machine Learning, including data types, control flow, functions, and basic data handling.


### **1. Import Libraries**

```python
import numpy as np
import pandas as pd
```

### **2. Python Data Types**

```python
# Numeric types
x = 10
y = 3.5

# String
name = "Machine Learning"

# Boolean
is_learning = True

x, y, name, is_learning
```

### **3. Lists, Tuples, Dictionaries**

```python
# List
numbers = [1, 2, 3, 4, 5]

# Tuple
dimensions = (1920, 1080)

# Dictionary
student = {
    "name": "Sinthol",
    "field": "CSE",
    "goal": "Machine Learning Engineer"
}

numbers, dimensions, student
```

### **4. Control Flow**

```python
score = 85

if score >= 80:
    result = "Excellent"
elif score >= 60:
    result = "Good"
else:
    result = "Needs Improvement"

result
```

### **5. Loops**

```python
for i in range(1, 6):
    print("Iteration:", i)
```

### **6. Functions**

```python
def calculate_accuracy(correct, total):
    return correct / total

calculate_accuracy(85, 100)
```

### **7. NumPy Basics**

```python
arr = np.array([10, 20, 30, 40])
arr.mean(), arr.max()
```

### **8. Pandas Basics**

```python
data = {
    "Model": ["Random Forest", "XGBoost", "SVM"],
    "Accuracy": [0.85, 0.88, 0.82]
}

df = pd.DataFrame(data)
df
```
